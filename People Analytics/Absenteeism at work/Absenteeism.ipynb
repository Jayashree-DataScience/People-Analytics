## Absenteeism at Work

## Problem 
Business enviornment today is highly competitiveness that leads to increased pressure and raised stress levels. It may result in minor illness. Employee may develop long term condition (ex:depression). Here, in this project, the objective is to predict whether the employees is going to be absent for specific duration (days /weeks etc). This analysis helps the business to reorganize the workforce to avoid lack of productivity
### Absenteeism
#### Absenteeism is absence from work during normal working hours resulting in temporary incapacity to execute regular working activity

## Group the reasons for absence

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df=df.drop(['Reason for Absence'],axis=1)
```

```python
df.head()
```

1. Group 1: reason for absence (1 to 14)
2. Group 2: reason for absence (15 to 17)
3. Group 3: reason for absence (18 to 21)
4. Group 4: reason for absence (22 to 28)

```python
reason_type_1=reason_columns.loc[:,1:14].max(axis=1)  # max is required because among 1 to 14 one or more column could be 1 so to say that reason is one among in this group we need to make sure to assign this to 1 (ie max value)
reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
reason_type_4=reason_columns.loc[:,22:].max(axis=1)
```

## Concatenate columns

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df.head(3)
```
